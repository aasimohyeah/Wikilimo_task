```python
import matplotlib
import matplotlib.pyplot as plt
import random
import xarray
import numpy as np

data=xarray.open_dataset("download.nc") #opening the dataset downloaded using the above script
print(data)#print information about the dataset
```

```python
#plotting histograms for all the days (t2m)
days=len(data['time'])

for i in range(days):
    plt.hist(data.t2m[i,:,:],histtype='step',bins=100)
    plt.show()
```

```python
#plotting histograms for all the days (tp)
days=len(data['time'])
print(days)
for i in range(days):
    plt.hist(data.tp[i,:,:],histtype='step',bins=100)
    plt.show()
```

```python
#plotting histograms for all the days (swvl1)
days=len(data['time'])
print(days)
for i in range(days):
    plt.hist(data.swvl1[i,:,:],histtype='step',bins=100)
    plt.show()
```

```python
#selecting a random day from the dataset and plotting it
for i in data:
    day=random.randrange(data['time'].size) #a random day is selected
    val=data[i][day]
    val.plot(size=6)
```

```python
#missing value adjustment is done here for eac dataset by replacing all NaN values with the mean of that dataset

for i in data:
    print(i,'\n')
    mean=float(data[i].mean())
    median=float(data[i].median())
    print('Before nan adjustment')
    print('Mean',mean)
    print('Median',median)
    print('Standard Deviation',float(data[i].std()))
    print('Variance',float(data[i].var()),'\n')
    
    data[i]=data[i].fillna(mean) #replacing missing values with the mean in the dataarray
    
    print('After nan adjustment')
    print('Mean',mean)
    print('Median',median)
    print('Standard Deviation',float(data[i].std()))
    print('Variance',float(data[i].var()),'\n')
```
